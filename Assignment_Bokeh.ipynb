## Question 2 : What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

### Glyphs are the basic building blocks of visualizations in Bokeh. A Glyph is a visual shape or the marker that represents data points. Bokeh provides a wide range og glyphs, including lines,circles,squares,triangles, and more
### To add glyphs to a bokeh plot, you can use the glyph methods provided by the figure object. These methods take array of data for the x and y co-ordinates, as well as any additional attributes or properties that you want to set for the glyph.Following is the example how to ass glyphs to a bokeh plot.

```python
import bokeh.io
import bokeh.plotting 
from bokeh.plotting import figure,output_file,show
## To show the output in this notebook we use
bokeh.io.output_notebook()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
p = figure(title = "My Plot" ,width = 400,height = 400,x_axis_label = "X", y_axis_label = "Y")
p.circle(x,y,size = 10,color = "red",alpha = 1) ##aplha controls the transparency of the plot
p.line(x,y,line_width = 2,color = "blue")
show(p)
```

Question 3 : How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


Here is the example for the above question, that demonstrate the customizing the appearance of a bokeh plot by modifying the axes, legend and the background and grid lines.

```python
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, output_file,show
from bokeh.models import Range1d
bokeh.io.output_notebook()
x1 = [1,2,3,4,5]
y1 = [5,4,3,2,1]
y2= [1,1.5,3,6,5]
#creating a figure with a gray background and grid lines
p = figure(title = "The Plot", width = 400 , height = 400,background_fill_color = '#f0f0f0')
```

```python
p.grid.grid_line_color = "white"
#Set the X-axis label and range
p.xaxis.axis_label = "X"
p.x_range = Range1d(0,6)
#Set the Y-axis label and range
p.yaxis.axis_label = "Y"
p.y_range = Range1d(0,6)

#Add two line glyphs with the legend label
p.line(x,y1,legend_label = "Line 1",line_width=2,color = "red")
p.line(x,y2,legend_label = "Line 2",line_width=2,color = "blue")
#setting up the legend location
p.legend.location = "top_left"
p.legend.orientation="horizontal"
show(p)
```
