Q3. Write a Python function to calculate the probability density function of a normal distribution with 
given mean and standard deviation at a given point.

```python
from scipy.stats import norm

def calculate_normal_pdf(mean, std_dev, x):
    """
    Calculate the probability density function of a normal distribution at a given point.

    Parameters:
    mean (float): Mean of the normal distribution.
    std_dev (float): Standard deviation of the normal distribution.
    x (float): Point at which to calculate the PDF.

    Returns:
    float: Probability density function value at the given point.
    """
    pdf_value = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution
x_value = 0.5  # Point at which to calculate the PDF

pdf_at_x = calculate_normal_pdf(mean, std_dev, x_value)
print(f"PDF at x = {x_value}: {pdf_at_x}")
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 
and plot a histogram of the results using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 1000  # Sample size
p = 0.4   # Probability of success

# Generate random sample
random_sample = np.random.binomial(n, p, size=1000)

# Plot histogram
plt.hist(random_sample, bins=20, density=True, alpha=0.7, color='red')
plt.title('Histogram of Random Sample from Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution 
with given mean at a given point.

```python
from scipy.stats import poisson

def poisson_cdf(mean, point):
    cdf_value = poisson.cdf(point, mu=mean)
    return cdf_value

# Example usage
mean = 3.5
point = 2
cdf_at_point = poisson_cdf(mean, point)
print(f"CDF at point {point} for Poisson(mean={mean}): {cdf_at_point}")
```

Q7. How Binomial distribution different from Poisson distribution?

Binomial:-
 It is used to model the number of successes in a fixed number of independent Bernoulli trials (experiments with two possible outcomes: success or failure). Each trial has the same probability of success.
 It is characterized by two parameters: the number of trials (n) and the probability of success on each trial (p).and for unsuccess its q = 1-p
 Assumes a fixed number of trials with two possible outcomes on each trial. The trials are independent and the probability of success remains constant across all trials.
 The shape of the distribution becomes more symmetric as the number of trials increases.
Possion:-
It is used to model the number of rare events that occur in a fixed interval of time or space. It is often applied to situations where events happen randomly and independently, such as the number of phone calls received in a call center within a specific time period.
It is characterized by a single parameter (λ) which represents the average rate of events occurring in the given interval.
Assumes a rare event occurrence with a very large number of possible trials. Events are assumed to occur randomly and independently.
It is a skewed distribution, often with a longer tail to the right.

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the 
sample mean and variance.

```python
from scipy.stats import poisson
import numpy as np

def poisson_ans(mean, size_sample):
    sample = np.random.poisson(mean , size = size_sample)
    return sample

# Example usage
mean = 5
size_sample = 1000
value = poisson_ans(mean, size_sample)
sample_mean= np.mean(value)
sample_variance = np.var(value)

print(f"Sample mean: {sample_mean}")
print(f"Sample variance: {sample_variance}")
```

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

The least frequent data points in a normal distribution appear in the tails, which are the regions that are farthest from the mean. The majority of the data is concentrated around the mean, and the frequency of data points gradually decreases as you move away from the mean in either direction.
